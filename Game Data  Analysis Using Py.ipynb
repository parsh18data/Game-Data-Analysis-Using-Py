```python
# Create a scatter plot
plt.scatter(df['robots_owned'], df['ch_fights'])


plt.xlabel('robots_owned')
plt.ylabel('ch_fights')
plt.title('Scatter Plot: Relationship between robots_owned and ch_fights')

plt.show()
```

```python
#regression plot between robots owned and ch flights
df = pd.read_csv('03_Data.csv')
x_column = 'robots_owned'
y_column = 'ch_fights'

sns.regplot(x=x_column, y=y_column, data=df)


plt.title(f'Regression Plot between {x_column} and {y_column}')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.show()
```
